```python
import numpy as np

tamanho = 300

# Frequência das ondas. Um valor maior cria anéis mais próximos.
frequencia = 0.04
# -----------------------------------------

# Cria uma grade de coordenadas x e y
x = np.arange(0, tamanho)
y = np.arange(0, tamanho)
xx, yy = np.meshgrid(x, y)

# Calcula a distância de cada ponto (x,y) até a origem (0,0)
distancia = np.sqrt(xx**2 + yy**2)

# Calcula o valor de cada ponto usando a função cosseno da distância
# Isso cria o padrão de onda circular com valores entre -1.0 e 1.0
matriz_de_dados = np.cos(frequencia * distancia)

np.savetxt('circular.txt', matriz_de_dados, fmt='%.4f', delimiter=' ')
```

```python
#gera o density plot colorido

from pylab import imshow,show
from numpy import loadtxt
data = loadtxt("circular.txt",float)
imshow(data)
show()
```

```python
#density plot monocromático

from pylab import imshow,gray,show
from numpy import loadtxt
data = loadtxt("circular.txt",float)
imshow(data,origin="lower")
gray()
show()
```
